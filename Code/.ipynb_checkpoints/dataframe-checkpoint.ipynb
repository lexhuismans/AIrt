# DataFrames

```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv(r'../data/zeefdrukken_data.csv')
df = df.iloc[0:-7,1::]
```

```python
def is_framed(value):
    if value == 'Ingelijst':
        return 1
    else:
        return 0
        
df['Ingelijst'] = df['Ingelijst'].apply(is_framed)
```

```python
def get_sizes(size):
    if type(size) == 'str':
        dimensions = re.findall(r'[+-]?\d*\.\d+|\d+', size)
        height = float(dimensions[0])
        width = float(dimensions[1])
        print('Height: ', height, '    Width: ', width, '    Datatypes: ', type(height), '  ', type(width))
        return height, width
    else:
        print('Return None')
        return None, None
    
df[['Hoogte', 'Breedte']] = df['Afmetingen'].apply(get_sizes)
```
